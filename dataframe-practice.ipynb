# DataFrameのデータ抽出メモ

## DataFrame作成

````python
# データフレーム作成
sampleDf = pd.DataFrame({'A': ['apple', 'air', 'area'],
                         'B': ['base', 'bag', 'ball'],
                         'C': ['cap', 'collection', 'cover']},
                        index=['2020-10-12', '2020-10-13', '2020-10-14']
                       )
sampleDf
````

## サンプルデータ取得（MySQL）
※ 今回は主にこちらを使用

````python
# import
import pandas as pd
import pymysql
import sqlalchemy as sqa
# DB情報
from config import DATABASE

# DATABASE接続情報
DATABASE = 'mysql+pymysql://%s:%s@%s/%s?charset=utf8' % (
    DATABASE["USERNAME"],
    DATABASE["PASSWORD"],
    DATABASE["HOSTNAME"],
    DATABASE["DBNAME"],
) 
# 接続
engine = sqa.create_engine(DATABASE, echo=False)
query = "SELECT * FROM emps"
df = pd.read_sql(query, con = engine)
df
````

## 特定の列を取得する
カラム名（列名）を指定して、特定の列を抽出できる

````python
# "em_name"列を取得する
df["em_name"]
````

````python
# .(ドット)を利用しても同じ結果が得られる
df.em_name
````

## 特定の区間の行を抽出する

````python
# 1行目から3行目を抽出
df[1:3]
````

````python
# 先頭から3行目までを抽出
df[:3]
````

## locアトリビュートを使って特定の行・列を抽出する
locアトリビュートを利用して、ラベルに基づいて特定の行や列を抽出できる

````python
# 行名が"10"の列を抽出
df.loc[10]
````

````python
# 列 "em_no", "em_name"の2列を抽出
df.loc[:, ['em_no', 'em_name']]
````

````python
# 行名 = "2" ～ "5" の "em_no"列と"B"列を取得
df.loc['2':'5', ['em_no', 'em_name']]
````

## 行や列の位置を指定して行・列を取得する
ilocアトリビュートを用いて、行や列の位置に基づいて行・列を取得することができる<br>行や列は0行目・0列目から始まる点に注意する

````python
# 3行目を取得
df.iloc[3]
````

````python
# 1,2,4行目と0-2列目を取得
df.iloc[[1,2,4],[0,2]]
````

````python
# 1-3行目とすべての列を取得
df.iloc[1:3,:]
````

````python
# すべての行と1-3列目を取得
df.iloc[:,1:3]
````

````python
# 1行目の1列目の値を取得
df.iloc[1,1]
````

## 条件を指定して行・列を取得する
```True```または```False```を返す式を指定することで、特定の条件式に基づく行・列の取得ができる

````python
# "em_no"列の値が0より大きい行を取得
df[df.em_no > 0]
````

```isin()```メソッドと組み合わせて使うことで、複数の特定の値を含む行・列の抽出ができる

````python
# データフレームdf2を作成
df2 = sampleDf.copy()
df2['E'] = ['every', 'event', 'earth']
df2
````

````python
# "E"列に"event"を値に持つ行を抽出
df2[df2['E'].isin(['event'])]
````

## 特定の行を削除する

````python
# import
import pandas as pd
import numpy as np

# データフレームdfを作成
df = pd.DataFrame(np.random.randn(6,4), columns=list('ABCD'))
df
````

````python
# 行5を削除
df.drop(5)
````

````python
# 行3と4を削除
df.drop([3,4])
````

## 特定の列を削除する

````python
# 列Aを削除
df.drop("A", axis=1)
````

より簡単な方法として、Pythonの```del```ステートメントを利用する方法もある

````python
# 列Aを削除
del df['A']
df
````

# DataFrameのデータ抽出（選択）処理

以下のデータを対象に、次のセクション以降で具体的なデータ抽出方法を説明

````python
# import
import pandas as pd
import numpy as np

df = pd.DataFrame({'都道府県': ['千葉県', '三重県', '北海道', '神奈川県'],
                  'まいわし': [10, 20, 100, np.nan],
                  'かたくちいわし': [20, 30, 60, 60],
                  'あじ': [np.nan, np.nan, 100, 30]}
                 ).set_index('都道府県')
df
````

## DataFrame[]

````python
df['まいわし'] # Seriesになる
````

````python
df[['まいわし']] # DataFrameになる
````

````python
df[['まいわし', 'かたくちいわし']]
````

## DataFrame.loc

````python
df.loc['千葉県'] #Seriesになる
````

````python
df.loc[['千葉県']] #DataFrameになる
````

````python
df.loc[['三重県', '神奈川県'],:]
````
